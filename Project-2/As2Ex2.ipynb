```python
import pandas as pd
```

```python
df = pd.read_table("lpmc02.dat")
```

```python
df.columns, df.shape
```

```python
# 4 strata

S1 = df[(df.age <= 40) & (df.female == 1)]
S2 = df[(df.age > 40) & (df.female == 1)]
S3 = df[(df.age <= 40) & (df.female == 0)]
S4 = df[(df.age > 40) & (df.female == 0)]

strata = {"S1": S1.shape[0], "S2": S2.shape[0], "S3": S3.shape[0], "S4": S4.shape[0]}
popu_est = {"S1": 2599058, "S2": 1765143, "S3": 2676249, "S4": 1633263}

S = sum(list(strata.values()))
N = sum(list(popu_est.values()))

df_sw = pd.DataFrame(columns=list(strata.keys()), index=["Size", "Weight"])
for key,val in strata.items():
    df_sw.at["Size", key] = val
    df_sw.at["Weight", key] = S/val * popu_est[key]/N
    
# associate each weight with an individual n

S1["Weights"] = df_sw.loc["Weight", "S1"]
S2["Weights"] = df_sw.loc["Weight", "S2"]
S3["Weights"] = df_sw.loc["Weight", "S3"]
S4["Weights"] = df_sw.loc["Weight", "S4"]

df_new = pd.concat([S1, S2, S3, S4]).reset_index(drop=True)    
```
